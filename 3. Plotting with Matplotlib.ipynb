# **Plotting with Matplotlib**

```python
# Import all the necessary libraries first
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Make your datasets
x = np.linspace(0,10,100)
y = x**3
```

```python
# Make your plots
plt.plot(x,y,'red',linewidth=3,linestyle="--")
```

```python
# Lets put labels on them
plt.plot(x,y,'red',linewidth=3,linestyle="--") #Repeat the same line as the above
plt.title("Number of Iterations vs Time")
plt.xlabel("Time")
plt.ylabel('Number of Iterations')
```

## **Creating Subplots**

```python
fig = plt.figure(figsize=(10,5)) # 10 represents the length and 5 represents the width of the plot.
plt.subplot(1,3,1)               # the (1,3,1) can be considered as (row,colum,plot numbers)
plt.plot(x,y,"r--")
plt.title("Plot 1")
plt.subplot(1,3,2)
plt.plot(y*2,x**2,"g:")
plt.title("Plot2")
plt.subplot(1,3,3)
plt.plot(x,y,"purple")
plt.title("Plot 3")
```

```python
# Lets create one small plot inside one large plot

# plt.figure() creates an empty figure
# figsize can be defined
fig = plt.figure(figsize = (10,6))

# Now add dimensions using, fig.add_axes(left,bottom,width,height) for each figure
fig1 = fig.add_axes([0.1,0.1,0.8,0.8])
fig2 = fig.add_axes([0.25,0.5,0.3,0.2])

# Now that the plotting area/canvas is ready, you can start plotting as well as adding the axes and titles.
fig1.plot(x,y)
fig1.set_title("Large Plot")
fig1.set_xlabel("Time")
fig1.set_ylabel("Number of Iterations")

fig2.plot(x*2,y,"g:")
fig2.set_title("Small Plot")
fig2.set_xlabel("Time")
fig2.set_ylabel("Number of Iterations")
```

```python
# Another method to draw subplots is by using "fig,axes"
fig,axes = plt.subplots(nrows = 2, ncols = 2, figsize = (5,5), dpi = 100) # dpi is the resolution

# Plot #1 in the 1st row and 1st column
axes[0,0].plot(x,y,'r-')
axes[0,0].set_title('Plot 1')
axes[0,0].set_xlabel('Time')
axes[0,0].set_ylabel('Production 1')

# Plot #2 in the 1st row and 2nd column
axes[0,1].plot(y,x,'g:')
axes[0,1].set_title('Plot 2')
axes[0,1].set_xlabel('Time')
axes[0,1].set_ylabel('Production 2')

# Plot #3 is in the 2nd row and 1st column
axes[1,0].plot(y,x**2,'b--')
axes[1,0].set_title('Plot 3')
axes[1,0].set_xlabel('Time')
axes[1,0].set_ylabel('Production 3')

#Plot #4 is in the 2nd row and 2nd column
axes[1,1].plot(x,y**2,'y-')
axes[1,1].set_title('Plot 4')
axes[1,1].set_xlabel('Time')
axes[1,1].set_ylabel('Production 4')

# To better present the graphs without over-lapping, use "plt.tight_layout"
plt.tight_layout()
```
